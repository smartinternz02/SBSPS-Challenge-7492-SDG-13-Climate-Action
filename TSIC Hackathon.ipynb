```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Weather_Jan_to_Mar_2021.xlsx')
data.head()
```

### Data Cleaning

```python
df = data.drop(data.columns[[1,8,9,10,11,12,13,14]], axis=1)
df.head()
```

### District wise analysis

```python
df_districts = df.groupby('District', as_index = False).mean()
df_districts.head()
```

```python
plt.figure(figsize=(14,7))
all_districts = df_districts['District']
avg_temp = (df_districts['Min Temp (°C)'] + df_districts['Max Temp (°C)']) /2
avg_humidity = (df_districts['Min Humidity (%)'] + df_districts['Max Humidity (%)'])/2
plt.plot(all_districts, avg_temp, 'r')
plt.ylabel('Avg Temperature')
plt.xticks(rotation = 90)
plt.title('Average Temperatures in Districts of Telangana (Past 3 months)')
plt.show()
```

```python
plt.figure(figsize=(14,7))
avg_humidity = (df_districts['Min Humidity (%)'] + df_districts['Max Humidity (%)'])/2
plt.bar(all_districts, avg_humidity)
plt.ylabel('Avg Humidity')
plt.xticks(rotation = 90)
plt.title('Average Humidity in Districts of Telangana (Past 3 months)')
plt.show()
```

## Month wise analysis

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df_months = df.groupby('Month', as_index = False).mean()
df_months
```

```python
plt.figure(figsize=(10,10))
all_months = ['Jan', 'Feb', 'Mar']
avg_temp = (df_months['Min Temp (°C)'] + df_months['Max Temp (°C)']) /2
avg_humidity = (df_months['Min Humidity (%)'] + df_months['Max Humidity (%)']) /2
plt.bar(all_months, avg_temp)
plt.ylabel('Avg Temperature')
plt.title('Month wise average Temperatures')
plt.show()
```

```python
plt.figure(figsize=(10,8))
all_months = ['Jan', 'Feb', 'Mar']
avg_humidity = (df_months['Min Humidity (%)'] + df_months['Max Humidity (%)']) /2
plt.bar(all_months, avg_humidity)
plt.ylabel('Avg Humidity')
plt.title('Month wise average Humidity')
plt.show()
```

## Results

From the two kinds of analysis done(District wise/Monthly), we can get some insights into the recent climate conditions of telangana. 
1. The highest avg temperature is recorded in Khammam, lowest in Rangareddy. 
2. As expected, the highest humidity is also recorded in khammam(as the temperatures are high).
3. There is a sudden increase in avg temperature in the month of march.
4. The avg humidity is decresing as the months pass by.

## Conclusion

With the help of the following insights drawn from the above graphs we can come to a conclusion that:
1. During the winter season northern telangana districts have the least temperature compared to other districts(first graph). This trend is going to change as the summer starts, so districts like adilabad should expect higher rise in temperature in upcoming months.
2. From march onwards, temperatures start rising gradually. (Third graph)
3. As the winter is ceasing, the humidity levels in telangana are about to fall gradually. (Fourth graph)
